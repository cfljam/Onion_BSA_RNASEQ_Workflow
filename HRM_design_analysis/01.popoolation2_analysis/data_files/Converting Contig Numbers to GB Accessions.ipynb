Read accessions into a  dict

```python
import pandas as pd
acc_df=pd.read_csv('GBGJ01_accs',skiprows=2,header=None,names=['Contig_id','Accession'],sep="\t")
acc_df.head()
acc_dict=dict(zip(acc_df.Contig_id,acc_df.Accession))
```

Replace all instances of old contigs with their GB accessions

```python
out=open('All_bolt_assoc.cmh', 'w')
with open('All_bolt_assoc.cmh.bak','r') as f:
    for line in f:
        acc=line.split("\t")[0]
        rest="\t".join(line.split("\t")[1:])
        if acc_dict.has_key(acc):
            out.write(acc_dict[acc] +"\t" +rest)
        elif acc[:2]=="JR":
            out.write(acc  +"\t"+rest)  
        else:
            print acc
out.close()
```
