### Read the Tabulated Data from Popoolation2

```python
import pandas as pd
Bolt_assoc_contigs=pd.read_csv('Bolt_assoc_contigs.csv')
Bolt_assoc_contigs.head()
```

First get the variants on the contigs of interest

```python
target_contigs=Bolt_assoc_contigs[Bolt_assoc_contigs.sig_ACP267 & Bolt_assoc_contigs.sig_all & \
                                  Bolt_assoc_contigs.sig_NotACP267].chrom
target_contigs.head()
```

Copied data over to local drive

```python
!ls /Data/BSARNASEQ/
```

Form   bed file for all  contigs in the assembly

```python
!samtools faidx /Data/BSARNASEQ/CUDH2150_trinity.merged_0.98
!head /Data/BSARNASEQ/CUDH2150_trinity.merged_0.98.fai
```

Get the bedfile of target contigs only

```python
contig_df=pd.read_table('/Data/BSARNASEQ/CUDH2150_trinity.merged_0.98.fai',\
                        usecols=[0,2],\
                        names=['contig','end'])
contig_df['start']=1
contig_df=contig_df[['contig','start','end']]
contig_df.head()
```

```python
contig_df[contig_df.contig.isin(target_contigs)].to_csv('sig_contigs.bed',sep='\t',index=False,header=False)
```

### Convert vcf to gff

### Generate list of Bi-allelic SNP Targets 

```python
!vcftools --vcf sig_contigs.recode.vcf --remove-indels --recode  --recode-INFO-all --out sig_contigs.snps
```

```python
%run galaxy-pcr-markers/vcf2gvf.py sig_contigs.snps.recode.vcf sig_contigs.snps.recode.gff
!awk '!/^#/ {print $1":"$2":"$3":"$4}'   sig_contigs.snps.recode.gff > sig_contigs.snps.recode.targets
```

Get 5 random lines to test

```python
!gsort -R sig_contigs.snps.recode.targets | head -n 5 > test
```

```python
%run galaxy-pcr-markers/design_primers.py -i /Data/BSARNASEQ/CUDH2150_trinity.merged_0.98 \
-g sig_contigs.recode.gff -T test  -n 1 -u 
```

#### Design to a sample of  300 targets chosen at random

```python
!gsort -R sig_contigs.snps.recode.targets | head -n 300 > SNP_targets
%sc HRM_primer_output=python galaxy-pcr-markers/design_primers.py \
-i /Data/BSARNASEQ/CUDH2150_trinity.merged_0.98 \
-g sig_contigs.recode.gff -T SNP_targets  \
-n 1 -p 60 -P 120 -u
```

Load into pandas data frame and get number of rows

```python
import pandas as pd
import StringIO
HRM_primer_df=pd.read_table(StringIO.StringIO(HRM_primer_output),sep=' ')
len(HRM_primer_df)
```

#### Inspect distribution of Tm difference

```python
HRM_primer_df.hist('Tm_difference')
```

Tabulate Sets with higher Tm differences

```python
HRM_primer_df[HRM_primer_df.Tm_difference > 0.4]
```

Form a contig column and check number of unique contigs

```python
HRM_primer_df['Contig']=[X.split(":")[0] for X in HRM_primer_df.SNP_Target_ID]
HRM_primer_df.Contig.nunique()
```
